# Google Gemma LLMs 🌟

The new lightweight LLMs by google are open source, easy to use and available on huggingface.

Gemma comes in two sizes: 7B parameters, for efficient deployment and development on consumer-size GPU and TPU and 2B versions for CPU and on-device applications. Both come in base and instruction-tuned variants.

```python
from transformers import AutoTokenizer

model = "google/gemma-2b-it"
tokenizer = AutoTokenizer.from_pretrained(model)
```

```python
from transformers import pipeline
import torch

pipeline = pipeline(
    "text-generation",
    model=model,
    model_kwargs={"torch_dtype":torch.bfloat16},
    device="cuda"
)
```

```python
messages = [{
    "role": "user",
    "content": "Tell me a joke about spaceships."
}]
```

```python
prompt = pipeline.tokenizer.apply_chat_template(
    messages,
    tokenize=False,
    add_generation_prompt=True)
```

```python
print(prompt)
```

```python
output = pipeline(
    prompt,
    max_new_tokens = 256,
    add_special_tokens = True,
    do_sample=True,
    temperature = 0.7,
    top_k = 50,
    top_p = 0.95
)
```

```python
print(output[0]["generated_text"][len(prompt):])
```
